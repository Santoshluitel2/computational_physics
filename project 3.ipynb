```python
#COMPUTATIONAL PHYSICS PROJECT 2 SANTOSH LUITEL 07.10.2020
import numpy as np
import matplotlib.pyplot as plt
import math

# initialize variables
def myfunction(n,k,angle,v0):
    v=v0
    theta= math.pi/180 * angle
    vx=(v*math.cos(theta))
    vy=v*math.sin(theta)
    dt=0.0005 #time step
    t=0         
    g=9.8       
    x=0
    y=0
    xx=0
    yy=0
   
    
    time,xaxis,yaxis,velocity=[t],[x],[y],[v]
    VX,VY=[vx],[vy]
    xx_arr,yy_arr=[xx],[yy]
    
#EULER'S APPROXIMATION METHOD TO FIND THE NEXT VELOCITY AS INITIAL VELOCITY AND dv/dt is given
    while yy>=(-5):
        xx+= (v0*math.cos(theta))*dt
        vyy=v0*math.sin(theta) -g*t
        yy+=vyy*dt
        
        ax=-k* (v)**(n-1)*vx
        ay=-k* (v)**(n-1) *vy -g
        vx+= dt*ax
        vy+=dt*ay
        v=(vx**2 +vy**2)**0.5
        t+=dt
        x+=vx*dt
        y+=vy*dt -0.5*g*(dt**2)
        
        xx_arr.append(xx)
        yy_arr.append(yy)
    
        velocity.append(v)
        xaxis.append(x)   
        yaxis.append(y)
        VX.append(vx)
        VY.append(vy)
        time.append(t)
        
    plt.plot(time,yaxis,label='with drag force')
    plt.plot(time,yy_arr,label='without drag force')
    plt.xlabel('time',size='20')
    plt.ylabel('height',size='20')
    plt.legend()
    plt.grid()
  
    plt.figure(figsize=(10,6))
    plt.plot(xaxis,yaxis,label='with drag force')
    plt.plot(xx_arr,yy_arr,label='without drag force')
    plt.ylabel('HEIGHT',size='20')
    plt.xlabel('HORIZONTAL DISTANCE',size='20')
    plt.title('motion of a particle in 2D',size='20')
    plt.legend()
    plt.grid()
    
        



```

```python
#USING THE INITIAL GIVEN VALUES
print(myfunction(1,0.8,34,22))
```

```python
#using n=3/2 which gives k is equal to 0.17056
print(myfunction(3/2,0.17056,34,22))
```

```python
#using n=2 which gives k is equal to 0.03636
print(myfunction(2,0.03636,34,22))
```
