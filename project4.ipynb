```python
from numpy import empty, zeros , max, linspace, meshgrid
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
#jacobi method
m=10
vtop=100
vbottom=0
vleft=0
vright=0
xmin=0
xmax=1
dx=(xmax-xmin)/m
phi=zeros([m+1,m+1],float)
phiprime=empty([m+1,m+1],float)
xlist=linspace(xmin,xmax, m+1)
ylist=linspace(xmin,xmax, m+1)
phi[0,:]=vleft
phi[m,:]=vright
phi[0,:]=vbottom
phi[0,:]=vtop
delta=1.0
target=1e-4
k=1
while delta > target:
    for i in range(m+1):
        for j in range(m+1):
            if i==0 or i==m or j==0 or j==m:
                phiprime[i,j]= phi[i,j]
            else:
                phiprime[i,j] = (phi[i+1,j]+phi[i-1,j]+phi[i,j+1]+phi[i,j-1])/4
    delta=max(abs(phi-phiprime))
    phi,phiprime=phiprime,phi
    k+=1
print(k)
x,y=meshgrid(xlist, ylist)
angle=30
fig=plt.figure(figsize=(15,10))
ax=fig.add_subplot(111,projection='3d')
sur=ax.plot_surface(x,y,phi,cmap=cm.coolwarm,linewidth=0, antialiased= False)
ax.view_init(30,angle)
ax.set_title('Surface Plot ( Jacobi Method) ', y = 1.1,size='20')
ax.text(1.0,0.4, 155,'Number of iterations = {}'.format(k),size='20')
ax.set_xlabel('y',size='20')
ax.set_ylabel('x',size='20')
ax.set_zlabel('Potential[v]',size='20')
plt.draw()

```

```python
#SOR METHOD(the same code was repeated)
m=10
w=2/(1+np.pi/m)
vtop=100
vbottom=0
vleft=0
vright=0
xmin=0
xmax=1
dx=(xmax-xmin)/m
phi=zeros([m+1,m+1],float)
phiprime=empty([m+1,m+1],float)
xlist=linspace(xmin,xmax, m+1)
ylist=linspace(xmin,xmax, m+1)
phi[0,:]=vleft
phi[m,:]=vright
phi[0,:]=vbottom
phi[0,:]=vtop
delta=1.0
target=1e-4
k=1
while delta > target:
    for i in range(m+1):
        for j in range(m+1):
            if i==0 or i==m or j==0 or j==m:
                phiprime[i,j]= phi[i,j]
            else:
                phiprime[i,j] = ((phi[i+1,j]+phiprime[i-1,j]+phi[i,j+1]+phiprime[i,j-1])/4)*w +(1-w)*phi[i,j]
    delta=max(abs(phi-phiprime))
    phi,phiprime=phiprime,phi
    k+=1
print(k)
x,y=meshgrid(xlist, ylist)
angle=30
fig=plt.figure(figsize=(15,10))
ax=fig.add_subplot(111,projection='3d')
sur=ax.plot_surface(x,y,phi,cmap=cm.coolwarm,linewidth=0, antialiased= False)
ax.view_init(30,angle)
ax.set_title('Surface Plot ( SOR Method) ', y = 1.1,size='20')
ax.text(1.0,0.4, 155,'Number of iterations = {}'.format(k),size='20')
ax.set_xlabel('y',size='20')
ax.set_ylabel('x',size='20')
ax.set_zlabel('Potential[v]',size='20')
plt.draw()
```
