```python
# import (visual) library

import numpy as np
import matplotlib.pyplot as plt
```

```python
# initialize variables
y = 10000   #METERS
v = 0       #METER/SECONDS
a=0         #METER^2 / SECONDS
step=0.01   #STEP SIZE
t=0         # SECONDS
```

```python
#constants
G = 6.67e-11  # Gravitational Constant
M = 5.99e24   # Mass of Earth
R = 6.378e06   # Radius of Earth 
k = 3e-04     # Drag Constant
m = 1         # Mass of the falling body

g = G*M / R**2 #ACCELERATION DUE TO GRAVITY




```

```python
#EMPTY LIST
height,velocity,acceleration,drag_force,time,= [],[],[],[],[]

```

```python
#LOOP FOR EULER ALOGRITHM
while(y>=0):
    a=(g/((1 + y/R)**2))-((k*v**2)/m)
   
    drag=k*(v**2)
    v+=a*step
    y-=v*step
    t+=step
   
    acceleration.append(a)
    velocity.append(v)
    drag_force.append(drag)
    height.append(y)
    time.append(t)
   



    
   
    
```

```python
#PLOTTTING OF GRAPH
fig,ax1 = plt.subplots(figsize=(19,9))
plt.grid(True)


ax2 = ax1.twinx()
ax1.plot(height, velocity, 'g-')

ax2.plot(height,drag_force, 'b-')



ax1.set_xlabel('height',size=26)
ax1.set_ylabel('velocity', color='g',size='26')
ax2.set_ylabel('drag force', color='b',size='26')

plt.show()



```

```python
#PLOTTING OF GRAPH
fig,ax1 = plt.subplots(figsize=(19,9))
plt.grid(True)



ax2 = ax1.twinx()
ax1.plot(time,velocity, 'g-')
ax2.plot(time, drag_force, 'b-')

ax1.set_xlabel('time',size='26')
ax1.set_ylabel('velocity', color='g',size='26')
ax2.set_ylabel('drag force', color='b',size='26')

plt.show()

```

```python
#PLOTTING OF GRAPH
plt.figure(figsize=(19,9))
plt.plot(time,acceleration)
plt.xlabel('TIME',size=30)
plt.ylabel('',size=30)
plt.plot(time,drag_force)
plt.legend(('acceleration', 'Drag force', 'Total message length'),
           loc='best', shadow=True,prop={"size":30})
plt.grid()
plt.tick_params(axis='both', labelsize=25)
```

```python
#euler-richardson method

#using same initial variables and same constant value
y1 = 10000   # Initial height(in meters)
v1 = 0
a1=0
step1=0.1
t1=0

G = 6.67e-11  # Gravitational Constant
M = 5.99e24   # Mass of Earth
R = 6.37e06   # Radius of Earth 
k = 3e-04     # Drag Constant
m = 1         # Mass of the falling body

g = G*M / R**2


```

```python
#EMPTY LIST
height1,velocity1,acceleration1,drag_force1,time1, = [],[],[],[],[]
```

```python
#LOOP FOR MID POINT ALGORITHIM

while y1>=0:
    a1=(g/((1 + y1/R)**2))-((k*v1**2)/m)
    drag1=k*(v1**2)
    ymid=y1+v1*0.5*step1
    vmid= v1+ 0.5*a1*step1
    amid=(g/((1 + ymid/R)**2))-((k*vmid**2)/m)
    v1+=amid*step1
    y1-=vmid*step1
    t1+=step1
   
    acceleration1.append(a1)
    velocity1.append(v1)
    drag_force1.append(drag1)
    height1.append(y1)
    time1.append(t1)
  




```

```python
#PLOTTING GRAPH
fig,ax1 = plt.subplots(figsize=(19,9))
plt.grid(True)


ax2 = ax1.twinx()
ax1.plot(height1, velocity1, 'g-')
ax2.plot(height1,drag_force1, 'b-')

ax1.set_xlabel('height',size='26')
ax1.set_ylabel('velocity', color= 'g',size='26')
ax2.set_ylabel('drag force', color='b',size='26')


plt.show()

```

```python
#PLOTTING GRAPH
fig,ax1 = plt.subplots(figsize=(19,9))
plt.grid(True)

ax2 = ax1.twinx()
ax1.plot(time1,velocity1, 'g-')
ax2.plot(time1, drag_force1, 'b-')

ax1.set_xlabel('time',size='26')
ax1.set_ylabel('velocity', color='g',size='26')
ax2.set_ylabel('drag force', color='b',size='26')

plt.show()

```

```python
#PLOTTING GRAPH
plt.figure(figsize=(19,9))
plt.plot(time1,acceleration1)

plt.xlabel('TIME',size=30)
plt.ylabel('',size=30)

plt.plot(time1,drag_force1)

plt.legend(('acceleration', 'Drag force', 'Total message length'),
           loc='best', shadow=True,prop={"size":30})
plt.grid()
plt.tick_params(axis='both', labelsize=25)
```
